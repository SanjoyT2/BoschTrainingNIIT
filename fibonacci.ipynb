# Fibonacci Series in Python

This notebook demonstrates different approaches to generate and print the Fibonacci series.

## Method 1: Simple Loop Approach (Most Efficient)

This is the most efficient way to generate the Fibonacci series for larger values.

```python
# Simple loop approach
n = int(input("Enter the number of terms: "))

a, b = 0, 1
print("Fibonacci Series:")

for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

## Method 2: Using a List

This approach stores the series in a list and returns it.

```python
def fibonacci(n):
    """Generate Fibonacci series up to n terms"""
    fib_series = [0, 1]
    for i in range(2, n):
        fib_series.append(fib_series[i-1] + fib_series[i-2])
    return fib_series[:n]

# Test the function
n = 10
result = fibonacci(n)
print(f"Fibonacci Series (first {n} terms): {result}")
```

## Method 3: Recursive Approach

This is a classic recursive implementation. Note: It's less efficient for larger values due to repeated calculations.

```python
def fib_recursive(n):
    """Calculate the nth Fibonacci number using recursion"""
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

# Generate Fibonacci series
n = 8
print(f"Fibonacci Series (first {n} terms):")
for i in range(n):
    print(fib_recursive(i), end=" ")
```

## Method 4: Optimized Recursive with Memoization

This improves the recursive approach by caching previously calculated values.

```python
def fib_memo(n, memo={}):
    """Calculate Fibonacci number with memoization"""
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    
    memo[n] = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    return memo[n]

# Generate Fibonacci series
n = 15
print(f"Fibonacci Series (first {n} terms):")
for i in range(n):
    print(fib_memo(i), end=" ")
```
